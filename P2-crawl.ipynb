### making lists

```python
title_list = []
title_2_list = []
view_list = []
tag_list = [[], [], [], [], [], [], [], [], [], [], [], [], [], [], []]
```

### import modules

```python
import requests
import bs4 as bs
import pandas as pd
```

### url assignment

```python
url = "https://www.varzesh3.com/news/"
```

### crawling pages

```python
for pagenumber in range(2000000,2001000):
    page = url + str(pagenumber) + "/"
    resp = requests.get(page)
    print(pagenumber)
    html = bs.BeautifulSoup(resp.content)
    try:
        title = html.find("h1", class_="headline").get_text().strip("\u200c").strip("\r").strip("\n").strip("(عکس)")
        print(title)
        title_list.append(title)
    except:
        title_list.append("")
        print("Title Not Found")
    try:
        title_2 = html.find("p", class_="lead").get_text().strip("\u200c").strip("\r").strip("\n").strip("(عکس)")
        print(title_2)
        title_2_list.append(title_2)
    except:
        title_2_list.append("")
        print("2nd Title Not Found")
    try:
        view = html.find("div", class_="news-info")
        view = int(float(view.find_all("span")[2].get_text().strip("\u200c").strip("\r").strip("\n").strip("(عکس)").split()[0][:-1]) * 1000)
        print(view)
        view_list.append(view)
    except:
        view_list.append("")
        print("View Not Found")
    try:
        try:
            tag = html.find("div", class_="tags-news")
            tag = tag.find_all("div")
            tag_is_gotten = True
        except:
            tag_is_gotten = False
            for i in range(len(tag_list)):
                tag_list[i].append("")
            print("Tag Not Found")
        if tag_is_gotten == True:
            try:
                for i in range(len(tag)):
                    span = tag[i].find("span").get_text()
                    tag_list[i].append(span)
                    print(span)
            except:
                for i in range(len(tag)):
                    tag_list[i].append("")
            for i in range(-(len(tag_list)-len(tag)), -1):
                tag_list[i].append("")
    except:
        for i in range(-(len(tag_list)-len(tag)), -1):
            tag_list[i].append("")
        print("Tags Not Found")
```

### making generator & iterator for page numbers

```python
iterator = (i for i in range(len(title_list)))
page_number = [number + 2000001 for number in iterator]
```

### making dictionary

```python
keys = ["page number", "title", "2nd title", "view", "1st tag", "2nd tag", "3rd tag", "4th tag", "5th tag", "6th tag", "7th tag", "8th tag", "9th tag", "10th tag", "11th tag", "12th tag", "13th tag", "14th tag"]
values = [page_number, title_list, title_2_list, view_list, tag_list[0], tag_list[1], tag_list[2], tag_list[3], tag_list[4], tag_list[5], tag_list[6], tag_list[7], tag_list[8], tag_list[9], tag_list[10], tag_list[11], tag_list[12], tag_list[13]]
dictionary = dict(zip(keys, values))
```

```python
df = pd.DataFrame(dictionary)
df
```
